## Import functions

## Get data
First, read in a bunch of pickles that contain numpy-converted image. I'm only doing this because `cctx.python` does not have a good way of adding modules, and it certainly doesn't suppoert `skimage`. 

```python
import os
from cPickle import load

data_source = 'test_data_numpy'
files = os.listdir(data_source)
images = []
for f in files:
    im = load(open(os.path.join(data_source, f), 'rb'))
    images.append(clip_at_percentile(im, 99.9))
        
```

Now we have a list of clipped images, where we have set all pixels above the 99.9th percentile to the value at that percentile. This just deals with zingers/overloads. **Very important step, do not skip**

## Get segmentation for images
Why these params? See other notebook... I made a histogram of all test images, and intensities below 15 are clearly the beamstop. You might be able to tweak these further, but I think they're pretty good.

```python
segmentations = []
for im in images:
    segmentations.append(region_based_seg(im, low_thresh=15, high_thresh=35))
```

Let's visualize these:

```python
from auto_detect_beamstop import plot_segmentation

# Ploting Foo
import matplotlib.pyplot as plt
fig, axes = plt.subplots(len(images) // 4 + 1, 4, figsize=(20,40))
axes = axes.flatten()

for im, seg, ax in zip(images, segmentations, axes):
    plot_segmentation(im, seg, ax)
```

Before we pat ourselves on the back, we need to export these again!

```python
from cPickle import dump
import numpy as np
for file, seg in zip(files, segmentations):
    dump(np.array(~seg, dtype=int), open("segmentation-" + file, 'wb')) # 0=Background, 1=Beamstop
```
